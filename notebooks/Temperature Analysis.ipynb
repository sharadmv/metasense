```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
data = pd.read_csv("../data/all/B4.csv", index_col='time')
data.index = pd.to_datetime(data.index)
```

```python
data['temperature-F'] = data['temperature'] * 9.0/5.0 + 32
data['T-norm'] = (data['temperature-F'] - data['temperature-F'].mean()) / (data['temperature-F'].std())
data['epa-T-norm'] = (data['epa-temperature-F'] - data['epa-temperature-F'].mean()) / (data['epa-temperature-F'].std())
data['nc-T-norm'] = (data['nc'] - data['nc'].mean()) / (data['nc'].std())
```

```python
data.plot(y=['temperature-F', 'epa-temperature-F'])
plt.show()
```

```python
data.plot(y=['T-norm', 'epa-T-norm', 'nc-T-norm'])
plt.show()
```
